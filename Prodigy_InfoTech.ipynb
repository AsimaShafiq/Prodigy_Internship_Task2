```python
# Import necessary libraries:

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
# Loading Data:

data = pd.read_csv("/content/train.csv")
# Displaying first 5 rows of data:

print(data.head())
# Checking shape of data:

print(data.shape)
# Checking null values in data:

print(data.isnull().sum())
# Filling missing numerical values with median or mean:

data["Age"].fillna(data["Age"].median(),inplace = True)
# Filling missing Categorical values with "Unknown" and mode:

data["Cabin"].fillna("Unknown",inplace = True)
# Mode returns series with mode values---> it gives value with it's index; We need just value so wrote [0]:

data["Embarked"].fillna(data["Embarked"].mode()[0],inplace = True)
# Checking important info of data:

print(data.info())
print(data.describe())
print(data.isnull().sum())
# Checking duplicated data :

print(data.duplicated().sum())
# Setting figure size:

plt.figure(figsize = (6,4))
# Creating countplot to show Survival distribution-->How many people survived and how many not:

sns.countplot(x = "Survived",data = data,palette = "pastel")
# Title of countplot:

plt.title("Survived Distribution",fontsize = 14)
plt.xlabel("0 = Not Survived , 1 = Survived",fontsize = 12)
plt.ylabel("Count",fontsize = 12)
plt.show()
#  Creating countplot to show survival distribution against sex:

plt.figure(figsize = (6,4))
sns.countplot(x = "Sex",hue = "Survived",data = data,palette = "pastel")
plt.title("Survival By Gender",fontsize = 14)
plt.show()
# Creating countplot to show survival distribution against Passenger class:

plt.figure(figsize = (6,4))
sns.countplot(x = "Pclass",hue = "Survived",data = data,palette = "Set2")
plt.title("Survival By Passenger Class",fontsize = 14)
plt.show()

# Creating countplot to show survival distribution against Embarked:

plt.figure(figsize = (6,4))
sns.countplot(x = "Embarked",hue = "Survived",data = data,palette = "pastel")
plt.title("Survival By Embarked",fontsize = 14)
plt.show()
# Creating a Histogram to show Age Distribution of people in the data:

plt.figure(figsize = (6,4))
sns.histplot(data["Age"],bins = 30 ,kde = True,color = "purple")
plt.title("Age Distribution",fontsize = 14)
plt.show()
# Correlation matrix to show correlation between different columns:

corr_matrix = data.corr(numeric_only=True)
sns.heatmap(corr_matrix,annot = True,cmap = "coolwarm")
plt.title("Correlation matrix (Numeric Columns Only)",fontsize = 15,color = "green")
plt.show()





```
